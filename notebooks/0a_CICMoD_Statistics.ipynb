### Statistics for indices included in CICMoD data set

In this project we work with **C**limate **I**ndex **C**ollection based on **Mo**del **D**ata (CICMoD) data set (https://github.com/MarcoLandtHayen/climate_index_collection). Here, we show some statistics:

- Pairwise correlation coefficients
- Mean and standard deviation

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

### Load data

Load collection of climate indices directly from GitHub release.

```python
data_url = (
    "https://github.com/MarcoLandtHayen/climate_index_collection/"
    "releases/download/v2023.03.29.1/climate_indices.csv"
)
```

```python
climind = pd.read_csv(data_url)
climind = climind.set_index(["model", "year", "month", "index"]).unstack(level=-1)["value"]
climind
```

```python
# Plot pairwise correlation coefficients (FOCI):
fig, ax = plt.subplots(figsize=(9,7))
plt.title('Pairwise correlation of climate indices (FOCI)', fontsize=18)
sns.heatmap(
    climind.loc["FOCI"].corr(), 
    cmap="RdBu", 
    xticklabels=True, 
    yticklabels=True,
    center=0,
    vmin=-1,
    vmax=1,
)
```

```python
# Plot pairwise correlation coefficients (CESM):
fig, ax = plt.subplots(figsize=(9,7))
plt.title('Pairwise correlation of climate indices (CESM)', fontsize=18)
sns.heatmap(
    climind.loc["CESM"].corr(), 
    cmap="RdBu", 
    xticklabels=True, 
    yticklabels=True,
    center=0,
    vmin=-1,
    vmax=1,
)
```

```python
# Compute mean and standard deviation (FOCI):
print("FOCI mean:\n==========\n",climind.loc["FOCI"].mean())
print("\n\nFOCI std dev:\n=============\n",climind.loc["FOCI"].std())
```

```python
# Compute mean and standard deviation (CESM):
print("CESM mean:\n==========\n",climind.loc["CESM"].mean())
print("\n\CESM std dev:\n=============\n",climind.loc["CESM"].std())
```

### Sahel Rainfall (PREC_SAHEL)

Inspect behaviour of **PREC_SAHEL**:

- Look for **trend**.
- Check standard deviation for sliding window of specified length, to check, if we have **non-stationary** characteristics.

```python
# Define function to specify bar color, according to sign of values:
def bar_color(data,color_pos,color_neg):
    return np.where(data.values>0,color_pos,color_neg).T
```

```python
# Separate data by model:
climind_FOCI = climind.loc[('FOCI')]
climind_CESM = climind.loc[('CESM')]
```

```python
# Select only column 'PREC_SAHEL', which is used as target:
prec_sahel_FOCI = climind_FOCI.loc[:, climind_FOCI.columns == 'PREC_SAHEL']
prec_sahel_CESM = climind_CESM.loc[:, climind_CESM.columns == 'PREC_SAHEL']
```

```python
# Inspect PREC_SAHEL: Plot as bar plot (red / blue)
fig, axes = plt.subplots(2,1,figsize=[16,8])

# FOCI
axes[0].bar(
    x=range(len(prec_sahel_FOCI.values[:120,0])),
    height=prec_sahel_FOCI.values[:120,0],
    color=bar_color(prec_sahel_FOCI[:120],'r','b')[0],
    label="PREC_SAHEL (FOCI)",
    alpha=0.5
)

axes[0].legend()
axes[0].set_title('Monthly mean rainfall anomalies in the African Sahel region (10 yrs)', fontsize=20)
#axes[0].set_xlabel('Time [month]', fontsize=16)
axes[0].set_ylabel('norm. Anomaly', fontsize=16)

# CESM
axes[1].bar(
    x=range(len(prec_sahel_CESM.values[:120,0])),
    height=prec_sahel_CESM.values[:120,0],
    color=bar_color(prec_sahel_CESM[:120],'r','b')[0],
    label="PREC_SAHEL (CESM)",
    alpha=0.5
)
axes[1].legend()
axes[1].set_xlabel('Time [month]', fontsize=16)
axes[1].set_ylabel('norm. Anomaly', fontsize=16)

# horizontal line indicating the origin
axes[0].plot([0., 120], [0,0], "k--", linewidth=1)
axes[1].plot([0., 120], [0,0], "k--", linewidth=1)
```

```python
len(prec_sahel_CESM)
```

```python
## De-trending PREC_SAHEL time series using polynomial fit (2nd order):

# Create x index:
x_index_FOCI = np.array(range(len(prec_sahel_FOCI)))
x_index_CESM = np.array(range(len(prec_sahel_CESM)))

# Get trend parameters:
trend_param_FOCI = np.polyfit(x_index_FOCI, prec_sahel_FOCI, 2)
trend_param_CESM = np.polyfit(x_index_CESM, prec_sahel_CESM, 2)
# print(trend_param_FOCI)
# print(trend_param_CESM)

# Get trend:
FOCI_trend = (x_index_FOCI**2)*trend_param_FOCI[0]+x_index_FOCI*trend_param_FOCI[1]+trend_param_FOCI[2]
CESM_trend = (x_index_CESM**2)*trend_param_CESM[0]+x_index_CESM*trend_param_CESM[1]+trend_param_CESM[2]
```

```python
## Plot PREC_SAHEL time series and corresponding trend:

# FOCI:
plt.figure(figsize=(15, 4))
plt.plot(range(len(x_index_FOCI)), prec_sahel_FOCI, label='PREC_SAHEL (FOCI)')
plt.plot(range(len(x_index_FOCI)), FOCI_trend, linestyle='--', label='FOCI trend')
plt.legend()
plt.show()

# CESM:
plt.figure(figsize=(15, 4))
plt.plot(range(len(x_index_CESM)), prec_sahel_CESM, label='PREC_SAHEL (CESM)')
plt.plot(range(len(x_index_CESM)), CESM_trend, linestyle='--', label='CESM trend')
plt.legend()
plt.show()
```

```python
## Check standard deviation for PREC_SAHEL time series with sliding window of specified length:

# Set window length:
window_length = 60

# Get number of windows, corresponding to total length of time series and specified window length:
n_windows_FOCI = int(len(prec_sahel_FOCI)/window_length)
n_windows_CESM = int(len(prec_sahel_CESM)/window_length)

# Initialize storage for standard deviation:
std_FOCI = np.zeros(n_windows_FOCI)
std_CESM = np.zeros(n_windows_CESM)

## Loop over number of windows:

# FOCI:
for i in range(n_windows_FOCI):
    std_FOCI[i] = np.std(prec_sahel_FOCI.values[i*window_length:(i+1)*window_length])

# CESM:
for i in range(n_windows_CESM):
    std_CESM[i] = np.std(prec_sahel_CESM.values[i*window_length:(i+1)*window_length])
```

```python
## Plot standard deviation of PREC_SAHEL time series, divided into windows of specified length:

plt.figure(figsize=(15, 4))
plt.plot(std_FOCI, label='std. dev. PREC_SAHEL (FOCI)')
plt.plot(std_CESM, label='std. dev. PREC_SAHEL (CESM)')
plt.title('Standard deviation of PREC_SAHEL time series, divided into windows of length 60 months', fontsize=16)
plt.xlabel('Window number', fontsize=16)
plt.ylabel('Standard deviation', fontsize=16)
plt.legend()
plt.show()

```
